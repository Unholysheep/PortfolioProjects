**Задание 5. В двух столбцах есть пропущенные значения. Один из них — `days_employed`. Пропуски в этом столбце вы обработаете на следующем этапе. Другой столбец с пропущенными значениями — `total_income` — хранит данные о доходах. На сумму дохода сильнее всего влияет тип занятости, поэтому заполнить пропуски в этом столбце нужно медианным значением по каждому типу из столбца `income_type`. Например, у человека с типом занятости `сотрудник` пропуск в столбце `total_income` должен быть заполнен медианным доходом среди всех записей с тем же типом.**

```python
for t in data['income_type'].unique():
    data.loc[(data['income_type'] == t) & (data['total_income'].isna()), 'total_income'] = \
    data.loc[(data['income_type'] == t), 'total_income'].median()
```

### Категоризация данных

**Задание 16. На основании диапазонов, указанных ниже, создайте в датафрейме `data` столбец `total_income_category` с категориями:**

- 0–30000 — `'E'`;
- 30001–50000 — `'D'`;
- 50001–200000 — `'C'`;
- 200001–1000000 — `'B'`;
- 1000001 и выше — `'A'`.


**Например, кредитополучателю с доходом 25000 нужно назначить категорию `'E'`, а клиенту, получающему 235000, — `'B'`. Используйте собственную функцию с именем `categorize_income()` и метод `apply()`.**

```python
def categorize_income(income):
    try:
        if 0 <= income <= 30000:
            return 'E'
        elif 30001 <= income <= 50000:
            return 'D'
        elif 50001 <= income <= 200000:
            return 'C'
        elif 200001 <= income <= 1000000:
            return 'B'
        elif income >= 1000001:
            return 'A'
    except:
        pass
```

```python
data['total_income_category'] = data['total_income'].apply(categorize_income)
```

**Задание 18. Создайте функцию, которая на основании данных из столбца `purpose` сформирует новый столбец `purpose_category`, в который войдут следующие категории:**

- `'операции с автомобилем'`,
- `'операции с недвижимостью'`,
- `'проведение свадьбы'`,
- `'получение образования'`.

**Например, если в столбце `purpose` находится подстрока `'на покупку автомобиля'`, то в столбце `purpose_category` должна появиться строка `'операции с автомобилем'`.**

**Используйте собственную функцию с именем `categorize_purpose()` и метод `apply()`. Изучите данные в столбце `purpose` и определите, какие подстроки помогут вам правильно определить категорию.**

```python
def categorize_purpose(row):
    try:
        if 'автом' in row:
            return 'операции с автомобилем'
        elif 'жил' in row or 'недвиж' in row:
            return 'операции с недвижимостью'
        elif 'свад' in row:
            return 'проведение свадьбы'
        elif 'образов' in row:
            return 'получение образования'
    except:
        return 'нет категории'
```

```python
data['purpose_category'] = data['purpose'].apply(categorize_purpose)
```

#### 3.3 Есть ли зависимость между уровнем дохода и возвратом кредита в срок?

```python
# воспользуемся сформированными ранее категориями дохода

data.groupby('total_income_category')['total_income_category'].count()
```

Как можно заметить, в выборке снова наблюдается большой разброс в репрезентативности групп, отчего можно предположить, что сформировать корректные выводы по ним будет проблематично. Проверим эту гипотезу. 

```python
# воспользуемся другими методами формирования отчёта 
# формируем таблицу с помощью группировки и объединения

data_debt_sum = data.groupby(['total_income_category']).agg({'debt':'sum'})
data_debt_count = data.groupby(['total_income_category']).agg({'debt':'count'})
debt_to_income = data_debt_sum.merge(data_debt_count, on='total_income_category', how='left') 
debt_to_income['процент невыплаты долга'] = debt_to_income['debt_x'] / debt_to_income['debt_y'] * 100
debt_to_income.rename(columns={'debt_x': 'число задолженостей', 'debt_y' : 'число займов'}).sort_values('процент невыплаты долга', ascending=False)
```

**Вывод:** 

Несмотря на малую репрезентативность некоторых категорий, можно сделать следующие выводы: группы с самым высоким доходом - **A**, и самым низким - **E**, несут за собой примерно одинаковое количество заёмщиков и одинаковое количество должников, что делает зависимость невыплаты долга от уровня дохода потенциально незначимой. Вместе с тем:
* группа **C**, представляющая собой людей со средним доходом и являющаяся самой крупной, находится на втором месте по количеству должников.
* группы **B** и **D**, отвечающие за клиентов с уровнем дохода выше и ниже среднего соответственно, занимают нижние места таблицы и возвращают долг в срок чаще других.

Так как самыми надёжными заёмщиками являются люди с уровнем дохода ниже среднего, сложно сделать вывод о том, что уровень дохода напрямую влияет на выплату в срок: уместно рассматривать этот критерий вместе с другими - например, возрастом или наличием детей.

#### 3.4 Как разные цели кредита влияют на его возврат в срок?

```python
# используем сводную таблицу для формирования отчёта

data_pivot_1 = data.pivot_table(index=['purpose_category'], values='debt', aggfunc='sum')
data_pivot_2 = data.pivot_table(index=['purpose_category'], values='debt', aggfunc='count')
debt_to_purpose = data_pivot_1.merge(data_pivot_2, on='purpose_category', how='left')
debt_to_purpose['процент невыплаты долга'] = debt_to_purpose['debt_x'] / debt_to_purpose['debt_y'] * 100
debt_to_purpose.rename(columns={'debt_x': 'число задолженостей', 'debt_y' : 'число займов'}).sort_values('процент невыплаты долга', ascending=False)
```

**Вывод:** 

Категория людей, берущих кредит на операции с недвижимостью, является самой надёжной в отношении погашения долга в срок. Вместе с тем люди, обращающиеся за займом в целях связанных с образованием или транспортным средством, демонстрируют увеличенную тенденцию к просрочке выплат.
